```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()

ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color='green', marker='d', linewidth=2, label='Boston')
# Create labels for the x and y axes.
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
# Set the y limit between 0 and 45.
ax.set_ylim(0, 45)
# Create a title.
ax.set_title('PyBer Fare by Month')
# Add a grid.
ax.grid()
# Add the legend.
ax.legend()
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
# Add title
plt.title("PyBer Fare by Month")
# Add axes labels
plt.xlabel("Fares($)")
plt.ylabel("Date")
# Display legend
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="c", label="Chicago")
# Title
ax.set_title("PyBer Fare by Month")
# Axes labels
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
# Invert y axis
ax.invert_yaxis()
# Add legend
ax.legend()
```

```python
# Create scatterplot
plt.scatter(y_axis, x_axis, color='r', label='Chicago')
# Add legend
plt.legend()
# Title and axes labels
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')
# Limit x axis
plt.xlim(0, 50)
# Invert y axis
plt.gca().invert_yaxis()
```

```python
# Create bubble chart
plt.scatter(x_axis, y_axis, s=[i * 3 for i in y_axis])
```

```python
# Create object oriented scatter plot
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create object oriented bubble chart
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[5*i for i in y_axis], color='skyblue', alpha=0.8, edgecolors='black', linewidths=2, label='Boston')
# Add legend
ax.legend()
# Add title and axes
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
# Limit x axis
ax.set_xlim(0, 50)
# Invert y axis
ax.invert_yaxis()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Set pie slice to explode
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Increase size
plt.subplots(figsize=(8, 8))

# Create pie chart using matlab
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Set pie slices to explode
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Increase size
fig, ax = plt.subplots(figsize=(8, 8))

# Create pie chart using object oriented approach
ax.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%', shadow=True, startangle=95, counterclock=False)
plt.show()
```
